```python
### Part 1: Understanding Product Categories
#Business Question**: What are the most popular product categories on Amazon UK, and how do they compare in terms of listing frequency?

#1. **Frequency Tables**:
    #- Generate a frequency table for the product `category`.
frequency_table = df['category'].value_counts()
print(frequency_table)
print("\n")
    #- Which are the top 5 most listed product categories?
top_5_categories = df['category'].value_counts().head(5)
print("The 5 most listed product categories are:",top_5_categories)
```

```python
# Bar chat for the top 5 categories
#Method 1
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
top_5_categories .plot(kind='bar')
plt.title('Top 5 product categories')
plt.xlabel('Categories')
plt.ylabel('Number of products')
plt.show()

#Method 2
import seaborn as sns
sns.barplot(x=top_5_categories.index, y=top_5_categories.values, palette="Set3") 
plt.xticks(rotation=90) ;
```

```python
### Part 2: Delving into Product Pricing
#**Business Question**: How are products priced on Amazon UK, and are there specific price points or ranges that are more common?
#1. **Measures of Centrality**:
```

```python
#- Calculate the mean, median, and mode for the `price` of products.
mean_price = df['price'].mean()
print("Average of product prices:",mean_price)
print("\n")

median_price = df['price'].median()
print("Median of product prices:",median_price)
print("\n")

mode_price = df['price'].mode()[0]
print("Mode of product prices:",mode_price)
print("\n")

#What's the average price point of products listed? How does this compare with the most common price point (mode)?
#Average price: 89.24380943923663
#As the average price is higher than the mode price, it means that there are products that are expensive that impacts the average
```

```python
#2. **Measures of Dispersion**:
```

```python
#- Determine the variance, standard deviation, range, and interquartile range for product `price`.
variance_price = df['price'].var()
print("Variance of product prices:", variance_price)
print("\n")

standard_deviation_price = df['price'].std()
print("Standard deviation of product prices:", standard_deviation_price)
print("\n")

range_price = df['price'].max() - df['price'].min()
print("Range of product prices:", range_price)
print("\n")

interquartile_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)
print("Interquartile range of product prices:", interquartile_price)
print("\n")

max_price = df['price'].min()
print("The maximum price is:",max_price)
min_price = df['price'].max()
print("The minimum price is:",min_price)

#- How varied are the product prices? Are there any indicators of a significant spread in prices?

#Product price variance: 119445.48532256528
#This indicates a large dispersion of prices around the average (some are much more expensive or cheaper than others)

#Standard deviation of product prices: 345.60886175352226
#This indicates high price dispersion because product prices deviate widely from the average.

#Product price range: 100000.0
#This indicates that there is a significant difference between the lowest price and the highest price of the products.

#Product price interquartile range: 36.0
#This means that product prices in the middle of the distribution are not as dispersed as the variance and standard deviation suggest 
#(the majority of product prices are concentrated in a relatively narrow range, 
#but there are extreme values which increase the variance and the standard deviation).
```

```python
### Part 3: Unpacking Product Ratings
```

```python
#1. **Measures of Centrality**:
#Zero values have been removed so that the calculation results are not aberrant
mean_rating = df[df['stars'] != 0]['stars'].mean()
print("The mean rating is:",mean_rating)

median_rating = df[df['stars'] != 0]['stars'].median()
print("The median rating is:",median_rating)

mode_rating = df[df['stars'] != 0]['stars'].mode()[0]
print("The mode rating is:",mode_rating)

#How do customers generally rate products? Is there a common trend?
#How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

```

```python
#2. **Measures of Dispersion**:
    #- Determine the variance, standard deviation, and interquartile range for product `rating`.
    #- Are the ratings consistent, or is there a wide variation in customer feedback?

var_rating = df[df['stars'] != 0]['stars'].var()
print("The variance rating is:",var_rating)

std_rating = df[df['stars'] != 0]['stars'].std()
print("The standard deviation rating is:",std_rating)


interquartile_rating = df[df['stars'] != 0]['stars'].quantile(0.75) - df[df['stars'] != 0]['stars'].quantile(0.25)
print("The interquartile rating is:",interquartile_rating)
print("\n")
```

```python
#3. **Shape of the Distribution**:
    #- Calculate the skewness and kurtosis for the `rating` column.
    #- Are the ratings normally distributed, or do they lean towards higher or lower values?
skew_rating = df[df['stars'] != 0]['stars'].skew()
print("The skewness rating is:",skew_rating)

kurtosis_rating = df[df['stars'] != 0]['stars'].kurtosis()
print("The kurtosis rating is:",kurtosis_rating)
```
